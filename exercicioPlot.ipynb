Implementar um codigo que some duas listas de mesmo tamanho

```python
from random import randint #importa a funcao de numero aleatorios
import time
listaA = []
listaB = []

def somaListas(lista1, lista2):
  result = []
  for i in range(len(lista1)):
    result.append(lista1[i]+lista2[i])
  return result

def constroiLista(tamanho):
  lst = []
  for i in range(tamanho):
    lst.append(randint(1,tamanho+1))
  return lst

temposExec = []
for j in range(1,1001):
  listaA=constroiLista(j)
  listaB=constroiLista(j)
  ini = time.time_ns()
  somaListas(listaA,listaB)#QUANTO TEMPO EXECUTOU??
  fim = time.time_ns()
  temposExec.append(fim-ini)
```

```python
import matplotlib.pyplot as plt

plt.plot(temposExec,label='tempos de execucao')
plt.title('Tempos de execucao da soma de listas')
plt.xlabel('Tamanho da lista')
plt.ylabel('Tempo (ns)')
plt.legend()
plt.grid(True)
plt.show()
```
